```python
# Outliers
Кіл_сть_відгуків_в_гуглі = []
for Кіл_сть_відгуків_в_гуглі in df['Кіл_сть_відгуків_в_гуглі']:
    if Кіл_сть_відгуків_в_гуглі > (df['Кіл_сть_відгуків_в_гуглі'].mean() + 3*df['Кіл_сть_відгуків_в_гуглі'].std()):
        Кіл_сть_відгуків_в_гуглі = (dfm['Кіл_сть_відгуків_в_гуглі'].mean() + 3*df['Кіл_сть_відгуків_в_гуглі'].std())
    elif Кіл_сть_відгуків_в_гуглі < (df['Кіл_сть_відгуків_в_гуглі'].mean() - 3*df['Кіл_сть_відгуків_в_гуглі'].std()):
        Кіл_сть_відгуків_в_гуглі = (dfm['Кіл_сть_відгуків_в_гуглі'].mean() - 3*df['Кіл_сть_відгуків_в_гуглі'].std())
    Кіл_сть_відгуків_в_гуглі.append(Кіл_сть_відгуків_в_гуглі)
df['Кіл_сть_відгуків_в_гуглі'] = Кіл_сть_відгуків_в_гуглі

Средний_чек = []
for Средний_чек in df['Средний_чек']:
    if Средний_чек > (df['Средний_чек'].mean() + 3*df['Средний_чек'].std()):
        Средний_чек = (dfm['Средний_чек'].mean() + 3*df['Средний_чек'].std())
    elif Средний_чек < (df['Средний_чек'].mean() - 3*df['Средний_чек'].std()):
        Средний_чек = (dfm['Средний_чек'].mean() - 3*df['Средний_чек'].std())
    Средний_чек.append(Средний_чек)
df['Средний_чек'] = Средний_чек

Людинопотік = []
for Людинопотік in df['Людинопотік']:
    if Людинопотік > (df['Людинопотік'].mean() + 3*df['Людинопотік'].std()):
        Людинопотік = (dfm['Людинопотік'].mean() + 3*df['Людинопотік'].std())
    elif Людинопотік < (df['Людинопотік'].mean() - 3*df['Людинопотік'].std()):
        Людинопотік = (dfm['Людинопотік'].mean() - 3*df['Людинопотік'].std())
    Людинопотік.append(Людинопотік)
df['Людинопотік'] = Людинопотік

Церква = []
for Церква in df['Церква']:
    if Церква > (df['Церква'].mean() + 3*df['Церква'].std()):
        Церква = (dfm['Церква'].mean() + 3*df['Церква'].std())
    elif Церква < (df['Церква'].mean() - 3*df['Церква'].std()):
        Церква = (dfm['Церква'].mean() - 3*df['Церква'].std())
    Церква.append(Церква)
df['Церква'] = Церква
```

```python
# Metrics
def metrics(X_train, X_test, y_train, y_test, y_pred, model):
    RMSE = [mean_squared_error(y_train, model.predict(X_train),squared=False), mean_squared_error(y_test, y_pred, squared=False)]
    R2 = [model.score(X_train, y_train), model.score(X_test, y_test)]
    results = {'RMSE': RMSE, 'R2':R2}
    return pd.DataFrame(results, index=['Train', 'Test'])
```

```python
# Function Max Leaf
def max_leaf_nodes(X_train, X_test, y_train, y_test, n):
    mse_train = []
    mse_test = []
    for i in n:
        ct = DecisionTreeClassifier(max_leaf_nodes = i, random_state=10).fit(X_train, y_train)
        mse_train.append(mean_squared_error(y_train, ct.predict(X_train)))
        mse_test.append(mean_squared_error(y_test, ct.predict(X_test)))
    fig, ax = plt.subplots(figsize=(8, 4))
    ax.plot(n, mse_train, alpha=0.5, color='blue', label='train')
    ax.plot(n, mse_test, alpha=0.5, color='red', label='test')
    ax.set_ylabel("MSE")
    ax.set_xlabel("max_leaf_nodes")
```

```python
from sklearn.cluster import KMeans
wcss = []
for i in range(1, 11):
    kmeans = KMeans(n_clusters = i, init = 'k-means++', random_state = 0)
    kmeans.fit(df)
    wcss.append(kmeans.inertia_)

plt.plot(range(1, 11), wcss)
plt.title('The Elbow Method')
plt.xlabel('Number of clusters')
plt.ylabel('WCSS')
plt.show()
```

```python
from sklearn.metrics import silhouette_score

sil = []

for k in range(2, 6):
    kmeans = KMeans(n_clusters = k).fit(df)  
    preds = kmeans.fit_predict(df)
    sil.append(silhouette_score(df, preds, metric = 'euclidean'))


plt.plot(range(2, 6), sil)
plt.title('Silhouette Method')
plt.xlabel('Number of clusters')
plt.ylabel('Sil')
plt.show()

for i in range(len(sil)):
    print(str(i+2) +":"+ str(sil[i]))
```
